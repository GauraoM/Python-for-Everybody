```python
import socket

# Create a socket having internet protocol and socket type
mysocket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

mysocket.connect(('data.pr4e.org',80))

cmd = "GET http://data.pr4e.org/romeo.txt HTTP/1.0\n\n".encode()
# Get the index page through http version 1.1
mysocket.send(cmd)

while True:
  data = mysocket.recv(512)
  if (len(data) < 1):
    break
  print(data.decode())
mysocket.close() 
```

#### Urllib 
##### Urllib package is the URL handling module for python. It is much simpler than Socket.

```python
import urllib.request, urllib.parse, urllib.error

file = urllib.request.urlopen("http://data.pr4e.org/romeo.txt")

for line in file:
  print(line.decode().strip())
```

#### Beautiful Soup
##### It is used to scrap the data from the websites for further use.

```python
from bs4 import BeautifulSoup
import ssl

# Ignoring ssl certification
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

url = input("Enter - ")
# read the whole page
html = urllib.request.urlopen(url, context=ctx).read()
soup = BeautifulSoup(html, 'html.parser')

tags = soup("a") # Find all anchor tags & extracting urls
for tag in tags:
  print(tag.get('href',None))
```
